### 1) df.dropna(axis = 1, thresh)
- axis = 1이므로 열을 삭제 (디폴트는 행 삭제 : axis = 0)
- thresh에 입력된 숫자 이상 데이터가 존재하지 않으면 그 행을 삭제한다.
    - 예를 들어 thresh = 100 이면 전체 데이터 891개 중에서 결측치가 791개 이상이라서 남은 데이터가 100개보다 더 적으면 그 행을 drop하는 것

```python
#예시: deck를 삭제하고 싶다. -> 300개 이상의 Non-NaN을 보유한 컬럼만 남기기
df2 = df.dropna(axis = 1, thresh = 300)
print(len(df.columns), len(df2.columns))
```

deck는 Non-NaN이 203개뿐이다. 그래서 300개 이상 갖고있어야 한다고 하면 이렇게 drop되어 없어짐

### 2) 결측치가 특정 "비율" 이상으로 가지고 있어야만 남기기

```python
df.isna().sum()/len(df)
```

위는 각 열별 결측치의 비율이다. 근데 결측치가 너무 많은 열은 버리고 싶다.  
예를 들어 결측치 비율이 15%를 넘기면 그 열 자체를 drop하고 싶을 때는..

```python
[df.isna().sum()/len(df) > 0.15]
```

```python
# 15% 넘는 결측치만 갖고 있는 열들 
df.columns[df.isna().sum()/len(df) > 0.15]
```

```python
df.drop(columns = df.columns[df.isna().sum()/len(df) > 0.15], axis = 1)
```
