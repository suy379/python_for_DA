## [pandas] melt, pivot_table 활용하여 데이터프레임 가공하기
### 1. pd.melt  
`pd.melt(dataframe, id_vars, value_vars, var_name, value_name)`: 열을 행으로 위치를 바꿔 주는데 유용하다. 원래 길게 늘어진 여러 개의 열을 하나의 열 안에 다수의 행으로 넣고자 할 때 사용.  
`참고` id_vars, value_vars에 따른 value 값을 집계하는 기능은 없음! 
    - id_vars: 위치를 유지할 컬럼명(범주형)
    - value_vars: 행으로 바꿀 컬럼명(범주형)
         (id_vars 의 컬럼 외에 전부 행으로 바꿀 거라면 지정 안해도 됨)
    - var_name: 위에서 행이 되면서 새로 만들어진 열 이름 지정
    - value_name: 행이 되면서 새로 만들어진 value값의 열 이름 지정

```python
import pandas as pd
import seaborn as sns

# 데이터셋 따로 필요 (깃허브에서 다운)
pew = pd.read_csv('pew.csv')
pew.head()
```

religion 열은 종교이며, <$10k부터~ Don't know 까지는 소득수준에 대해서이다.  
각 종교 및 소득수준에 해당하는 사람들의 수를 카운트한 데이터이다.  
하지만, 소득수준 열이 너무 많지 않은가? (무려 10개..) 이러한 '넓은 데이터' 를 간단히 만들어보자

```python
# religion 열은 그대로 / 나머지 소득수준 열은 'income' 이라는 하나의 열에 모두 집어넣기
pd.melt(pew, id_vars = 'religion',
                       var_name = 'income',
                        value_name = 'cnt')
```

### 2. pd.pivot_table
`df.pivot_table( index, columns, values, aggfunc)` : 보통 table을 보기 좋은 형태로 가공한다면 pivot_table을 활용하는 경우가 대다수이다. 그만큼 중요!    
`참고` index, columns에 따른 values 값을 집계하기 위해선 aggfunc를 설정한다. 디폴트는 평균.  
    - index: 행 인덱스(범주형)
    - columns: 열 인덱스(범주형) 
    - values: 위의 피벗할 열의 데이터가 될 열 (수치형)
    - aggfunc: values를 집계할 때 사용할 함수 지정

```python
# seaborn에서 제공하는 데이터 사용
dia = sns.load_dataset('diamonds')
dia.head()
```

데이터셋 설명  
- carat: 다이아몬드의 무게
- cut: 잘려진 상태
- color: 색상
- clarity: 투명도 (I1: worst, SI2, SI1, VS2, VS1, VVS2, VVS1, IF: best)
- depth: 총 깊이(%) x, y, z값으로 산정
- table: 가장 넓은 지점에 대한 다이아몬드 상단의 너비
- price: 다이아몬드 가격
- x: 길이 (mm)
- y: 너비 (mm)
- z: 깊이 (mm)
#### Q1. 다이아몬드의 cut과 color 상태에 따른 가격(price)의 평균은?

```python
dia_pivot1 = pd.pivot_table(dia, index = 'cut',
                                  columns = 'color',
                                  values = 'price') #aggfunc = 'mean' 지정 안해도 됨(디폴트)
dia_pivot1
```

```python
# 피벗테이블의 좋은점 -> 이 상태로 바로 히트맵 만들기도 가능
sns.heatmap(dia_pivot1)
```

색이 옅어질수록 평균가격이 높다는 뜻.  
즉 cut 상태가 Premium일수록, color가 I, J에 가까울수록 가격이 높아짐  
#### Q2. 다이아몬드의 cut, color를 행인덱스 / clearity를 열인덱스로 하여 중위가격을 구하면?

```python
import numpy as np
dia_pivot2 = pd.pivot_table(dia, index = ['cut', 'color'],
                          columns = 'clarity',
                           values = 'price',
                           aggfunc = np.median)
dia_pivot2
```

```python
# 피벗테이블은 자동으로 인덱스가 설정됨. 인덱스 삭제 
dia_pivot2.reset_index(inplace = True)
dia_pivot2
```

```python
# 만들어진 것을 보니 다시 열이 길어진 '넓은 데이터'가 되었다. 열 'clarity'로 다시 만들기! 
pd.melt(dia_pivot2, id_vars = ['cut', 'color'],
                                    var_name = 'clarity',
                                   value_name = 'med_value')
```

#### Q3. 다이아몬드의 cut, color에 따라, 가격(price)의 중위수 & 무게(carat)의 합계가 궁금하다.

```python
dia.head(3)
```

```python
import numpy as np
pd.pivot_table(dia, index = 'cut',
                                  columns = 'color',
                                  values = ['price', 'carat'],
                                  aggfunc = {'price': np.median,
                                                        'carat': np.sum})
```

```python
# 넘파이 사용하지 않아도 됨 
pd.pivot_table(dia, index = 'cut',
                                  columns = 'color',
                                  values = ['price', 'carat'],
                                  aggfunc = {'price': 'median',
                                                        'carat': 'sum'})
```
